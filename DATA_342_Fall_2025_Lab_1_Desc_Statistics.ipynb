Step 4: Load the Excel File


```python
import requests
import pandas as pd

# Replace 'YOUR_FILE_ID_HERE' with the actual file ID from your Google Sheets share link
file_id = '154Hbo1On0Sh9tFl-lEt0bY7HBbCa6_-6'
url = f'https://docs.google.com/spreadsheets/d/{file_id}/export?format=xlsx'
response = requests.get(url)

if response.status_code == 200:
    with open('DATA 342 Fall 2025 Lab 1.xlsx', 'wb') as f:
        f.write(response.content)
    df = pd.read_excel('DATA 342 Fall 2025 Lab 1.xlsx', sheet_name='Daily Returns 2024', engine='openpyxl')
    df.columns = df.columns.str.replace('\xa0', ' ', regex=False).str.strip()
    df['Daily Return'] = pd.to_numeric(df['Daily Return'], errors='coerce')
    print("File downloaded and loaded successfully!")
else:
    print(f"Error downloading file: {response.status_code}")
    print("Please ensure the Google Sheet is shared publicly or with 'Anyone with the link' having 'Viewer' access.")
```

Step 6: Clean and Format the Data


```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.dropna(subset=['Daily Return'])  # Remove rows with missing returns
```
